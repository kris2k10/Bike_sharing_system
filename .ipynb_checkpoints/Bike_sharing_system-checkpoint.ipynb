# Assignment : Bike sharing system

A bike-sharing system is a service in which bikes are made available for shared use to individuals on a short term basis for a price or free. Many bike share systems allow people to borrow a bike from a "dock" which is usually computer-controlled wherein the user enters the payment information, and the system unlocks it. This bike can then be returned to another dock belonging to the same system.

A US bike-sharing provider BoomBikes has recently suffered considerable dips in their revenues due to the ongoing Corona pandemic. The company is finding it very difficult to sustain in the current market scenario. So, it has decided to come up with a mindful business plan to be able to accelerate its revenue as soon as the ongoing lockdown comes to an end, and the economy restores to a healthy state. 

As a part of this exercise we are required to model the demand for shared bikes with the available independent variables. It will be used by the management to understand how exactly the demands vary with different features. They can accordingly manipulate the business strategy to meet the demand levels and meet the customer's expectations. Further, the model will be a good way for management to understand the demand dynamics of a new market.

Model Building

In the dataset provided, there are three columns named 'casual', 'registered', and 'cnt'. The variable 'casual' indicates the number casual users who have made a rental. The variable 'registered' on the other hand shows the total number of registered users who have made a booking on a given day. Finally, the 'cnt' variable indicates the total number of bike rentals, including both casual and registered. The model should be built taking this 'cnt' as the target variable.

### Data Dictionary
- instant : Record index, we can safely drop this column. 
- dteday  : date
- season  : season (1:spring, 2:summer, 3:fall, 4:winter)
- yr      : year (0: 2018, 1:2019)
- mnth    : month ( 1 to 12)
- holiday : weather day is a holiday or not
- weekday : day of the week
- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.
- weathersit : 
	- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
	- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
	- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- temp   : temperature in Celsius
- atemp  : feeling temperature in Celsius
- hum    : humidity
- windspeed : wind speed
- casual : count of casual users
- registered: count of registered users
- cnt    : count of total rental bikes including both casual and registered

## Outliers detection treatment

### Using IQR ( inter quantile range) method

```python
### Variable having outliers are "windspeed" and "hum"
def treat_outlier(col):
    q1,q3 = col.quantile([0.25,0.75])
    print("First Quartlie- ", round(q1,2))
    print("Third Quartlie- ", round(q3,2))
    IQR = round(q3-q1,2)
    print(IQR)
    lower_range = q1-(1.5*IQR)
    upper_range = q3 + (1.5*IQR)
    return lower_range, upper_range
```

```python
l_windspeed,u_windspeed = treat_outlier(bike.windspeed)
print('Lower quantile range :{:.2f} and upper quantile range {:.2f}'.format(l_windspeed,u_windspeed))
```

```python
bike[bike.windspeed >= u_windspeed]
```

Observation: There are 13 records coming as outliers, but these seems legitimate transactions. So let's try with Z score technique to remove outliers

## Chi square test of independence on "holiday" and "workingday" columns  

### Hypothesis
 - H0(null hypothesis) : "holiday" and "workingday" are independent
 - H1(alternate hypothesis) : "Holiday" and "workingday" are dependent to each other
 - if p value is significent(p<alpha), then you can reject the null hypothesis. 

```python
bike.head(2)
```

```python
crosstab = pd.crosstab(bike['holiday'],bike['workingday'])
crosstab
```

```python
from scipy import stats
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(crosstab)
print("Chi-square value: ", chi2)
print("p-value: ", p)
print("Degrees of freedom: ", dof)
print("Expected Values: \n", expected)
```

## Observation:
 - Since p ~ 0, that means we can reject the null hypothesis and conclude that both variable are dependent to each other. 

###  1. ANOVA test for season and count of bike rent
 - H0(null hypothesis) : The average number of bike rented is same for all season
 - H1(alternate hypothesis) : The average number of bike rented is different for atleast one season. 


```python
from statsmodels.formula.api import ols                      # For n-way ANOVA
from statsmodels.stats.anova import _get_covariance,anova_lm # For n-way ANOVA
```

```python
# Fit the function
formula = 'cnt ~ C(season)'
model = ols(formula, bike).fit()

# Get ANOVA table
aov_table = anova_lm(model)
aov_table
```

**Conclusion: Since the p value is less than the significance level (0.05), we can reject the null hupothesis and conclude that
the bike rental differs for atleast one season)**

###  2. ANOVA test for weathersit and count of bike rent

```python
# Fit the function
formula = 'cnt ~ C(weathersit)'
model = ols(formula, bike).fit()

# Get ANOVA table
aov_table = anova_lm(model)
aov_table
```

**Conclusion: Since the p value is less than the significance level (0.05), we can reject the null hupothesis and conclude that
the bike rental differs for atleast one weathersit)**

## Observation:

 - Case 1: high VIF, high p-value => variable is very well explained by other features also non significent. 
 - Case 2: 
      - Low VIF, High p-Value => check by dropping this first. 
      - High VIF, low p-value
 - Case 3: Low VIF, Low p-value => not good for dropping.
